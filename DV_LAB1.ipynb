```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data.csv' is your dataset file
# Replace this with the actual path to your dataset
# Use pd.read_csv() or pd.read_excel() for different file types
data = pd.read_csv('/content/spotify-2023.csv', encoding='ISO-8859-1')
```

```python
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style and color palette
sns.set(style='whitegrid')
sns.set(palette="Set1")

# Scatter Plot 1: BPM vs Energy
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data['bpm'], y=data['energy_%'], color='blue', marker='o')
plt.title('BPM vs Energy', fontsize=16)
plt.xlabel('BPM', fontsize=14)
plt.ylabel('Energy %', fontsize=14)
plt.show()

# Scatter Plot 2: Speechiness vs Danceability
plt.figure(figsize=(6,4))
sns.scatterplot(x=data['speechiness_%'],y=data['danceability_%'],color='red',marker='^')
plt.title("Speechiness vs Danceablity",fontsize=16)
plt.xlabel('Speechiness',fontsize=14)
plt.ylabel('danceablity',fontsize=14)
plt.show()


# Scatter Plot 3: Spotify Playlists vs Danceability
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data['in_spotify_playlists'], y=data['danceability_%'], color='red', marker='s')
plt.title('Spotify Playlists vs Danceability', fontsize=16)
plt.xlabel('Spotify Playlists', fontsize=14)
plt.ylabel('Danceability %', fontsize=14)
plt.show()

# Scatter Plot 4: Released Month vs Apple Playlists
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data['released_month'], y=data['in_apple_playlists'], color='blue', marker='D')
plt.title('Released Month vs Apple Playlists', fontsize=16)
plt.xlabel('Released Month', fontsize=14)
plt.ylabel('Apple Playlists', fontsize=14)
plt.show()

# Scatter Plot 5: Released Year vs Spotify Charts
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data['released_year'], y=data['in_spotify_charts'], color='orange', marker='^')
plt.title('Released Year vs Spotify Charts', fontsize=16)
plt.xlabel('Released Year', fontsize=14)
plt.ylabel('Spotify Charts', fontsize=14)
plt.show()
```
